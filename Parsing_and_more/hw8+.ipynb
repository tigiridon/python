```python
import numpy as np
import pandas as pd
#from datetime import datetime
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('sberdata.csv', encoding = 'ANSI')
df.head()
```

```python
regions = set(df['region'].values)
regions
```

```python
params = set(df['name'].values)
params
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df.head()
```

```python
# выбираем параметры для выбора данных
region = 'Россия'
param = 'Средняя пенсия'
start_period = '2016-01-01'
end_period = '2018-12-31'
```

```python
df1 = df[(df['region']==region) & (df['name'] == param) & (df['date']>=start_period) & (df['date']<=end_period)]
df1
```

```python
df1.plot(x = 'date', y = 'value', figsize = (16,6), title = param)
```

```python
yr = set(df1['year'].values)
ax = plt.gca()
for y in yr:
    df1[df1['year'] == y].plot(x = 'month', y = 'value' , title = param, figsize = (16,6), label = y,ax=ax)
plt.show()
```

```python
regions = ['Россия',
           'Калининградская область',
           'Москва',
           'Ямало-Ненецкий автономный округ']
```

```python
df2 = df[(df['name'] == param) & (df['date']>=start_period) & (df['date']<=end_period)]
ax = plt.gca()
for reg in regions:
    df2[df2['region'] == reg].plot(x = 'date', y = 'value', figsize = (16,6), title = param, label = reg, ax = ax)
plt.show()
```
