```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import math
```

# 2

```python
def equations(p):
    x, y, l = p
    return (4*x+12*y+2*l*x, 64*y + 12*x + 32*l*y, x*x+16*y*y-64)

for l_i in range(-100,100,1):
  x, y, l =  fsolve(equations, (1, 1, l_i))
  print(l_i, x, y, l)
```

```python

def equations(p):
    x, y = p
    return (x*x - y*y + 3*x*y*y*y - 2*x*x*y*y+2*x-3*y-5, 3*y*y*y-2*x*x+2*x*x*x*y-5*x*x*y*y+5)
for x_i in range(-100,100,1):
  for y_i in range(-100,100,1):
    x, y =  fsolve(equations, (x_i, y_i))
    print(x_i, y_i, x, y)
```

```python
x = np.linspace(0, 20, 10000)
Q = [0]
Z = [1]

d = 0.002
x_root = []
eq = True

for i in range(0, len(x)-1):
  Q.append(Q[i]+Z[i]*(x[i+1]-x[i])/2)
  Z.append(Z[i]+(2-2*Q[i])*(x[i+1]-x[i])/2) 
  if x_root:
    eq = not (x_root[-1]+1)>x[i]
  if ((abs(Q[i]-Z[i])) < d) and eq:
    x_root.append(x[i])

plt.plot(x,Q)
plt.plot(x,Z)

print(f'korni:', *x_root)

plt.show()
```
