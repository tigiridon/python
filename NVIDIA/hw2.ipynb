## Задача № 1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.

●	Игнорируем слова, частота которых в документе строго превышает порог 0.9 с помощью max_df.

●	Ограничим количество слов, попадающий в мешок, с помощью max_features = 1000.

●	Исключим стоп-слова с помощью stop_words='english'.

●	Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с помощью CountVectorizer.get_feature_names().


```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(max_df = 0.9, max_features = 1000, stop_words='english')
```

#### CountVectorizer - tweet_stemmed

```python
# Создаем the Bag-of-Words модель колонки tweet_stemmed
bag_of_words = count_vectorizer.fit_transform(tweet_stemmed)
```

```python
# Отобразим Bag-of-Words модель как DataFrame
feature_names = count_vectorizer.get_feature_names()
CountVectorizer_cts = pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
CountVectorizer_cts.head(8)
```
